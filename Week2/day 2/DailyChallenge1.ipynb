```python
from PIL import Image
import matplotlib.pyplot as plt
# Load an image using PIL
image_path = '/content/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
from scipy.ndimage import rotate
def rotate_image_90_degrees(image):
  return rotate(image, 90, reshape=False)
```

```python
rotated_image = rotate_image_90_degrees(original_image)
plt.imshow(rotated_image)
```

```python
from PIL import ImageOps

def flip_image(image):
  image_mirrored = ImageOps.mirror(image)
  return ImageOps.flip(image_mirrored)
```

```python
mirrored_pliped_image = flip_image(original_image)
plt.imshow(mirrored_pliped_image)
```

```python
def resize_image(image, size):
  width, height = image.size
  new_width = int(width * size)
  new_height = int(height * size)
  return image.resize((new_width, new_height))
```

```python
resized_image = resize_image(original_image, 1.2)
plt.imshow(resized_image)
```
