ðŸŒŸ Exercise 5: Comparing Anxiety Levels Across Different Genders

```python
!unzip "/content/drive/MyDrive/GENAI/Week2/Day4/student mental health.zip"
```

```python
mental_health_file = pd.read_csv("/content/Student Mental health.csv")
df = pd.DataFrame(mental_health_file)
df.head()
```

```python
df_test = df.groupby("Choose your gender").agg({"Do you have Anxiety?" : "count"})
df_test
```

```python

sns.barplot(data = df_test, x = "Choose your gender", y= "Do you have Anxiety?")
plt.title("Anxiety Levels Across Different Genders")
plt.xlabel("Gender")
plt.ylabel("Anxiety")
plt.show()
```
