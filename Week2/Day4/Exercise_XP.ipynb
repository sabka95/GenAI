ðŸŒŸ Exercise 3: Visualizing Monthly Sales with a Bar Chart

```python
month_sales = [5000, 5500, 6200, 7000, 7500, 6500, 6800, 5900, 4000, 5700, 9000, 8300]
monthes = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

fig = plt.bar(monthes, month_sales)
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly sales")
plt.show()
```

ðŸŒŸ Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
df.head()
```

```python
df["Do you have Panic attack?"] = df["Do you have Panic attack?"].map({"Yes" : 1, "No" : 0})
df.head()
```

```python
sns.scatterplot(data = df, x = "Age", y = "Do you have Panic attack?", marker="*", color = "green")
plt.title("Age vs Panic Attacks")
plt.xlabel("Age")
plt.ylabel("Panic Attacks")
plt.show()
```
