```python
import pandas as pd

salaries_files = pd.read_csv("/content/Data Science Job Salary dataset/datascience_salaries.csv")
salaries_df = pd.DataFrame(salaries_files)
salaries_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

salaries_df_temp = salaries_df.copy()

scaler = MinMaxScaler()
salaries_df_temp["salary_narmalize"] = scaler.fit_transform(salaries_df_temp[["salary"]])
salaries_df_temp = salaries_df_temp.drop(columns = ["salary"])
salaries_df_temp.head()
```

```python
from sklearn.preprocessing import LabelEncoder

salaries_df2 = salaries_df_temp.copy()

le = LabelEncoder()
salaries_df2["salary_currency"] = le.fit_transform(salaries_df2["salary_currency"])
salaries_df2["location"] = le.fit_transform(salaries_df2["location"])
salaries_df2["experience_level"] = le.fit_transform(salaries_df2["experience_level"])
salaries_df2["job_type"] = le.fit_transform(salaries_df2["job_type"])
salaries_df2["job_title"] = le.fit_transform(salaries_df2["job_title"])
salaries_df2.head()
```

```python
from sklearn.decomposition import PCA

numeric_cols = salaries_df2.select_dtypes(include=['number']).columns

pca = PCA(n_components=2)
reduced_salaries = pca.fit_transform(salaries_df2[numeric_cols])
reduced_salaries_df = pd.DataFrame(reduced_salaries, columns=["PC1", "PC2"])
reduced_salaries_df.head()
```

```python
mean_salaries = salaries_df.groupby("experience_level")["salary"].mean()
mean_salaries
```

```python
median_salaries = salaries_df.groupby("experience_level")["salary"].median()
median_salaries
```
