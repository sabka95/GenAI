ðŸŒŸ Exercise 6 : Basic Hypothesis Testing

```python
from scipy import stats

np.random.seed(42)

# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

t_statistic, p_value = stats.ttest_rel(productivity_before, productivity_after)


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in productivity.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in productivity.")
```

ðŸŒŸ Exercise 8 : Time Series Data Manipulation

```python
dates = np.arange('2023-01-01', '2023-12-31', dtype='datetime64[D]')
dates = pd.DataFrame(dates, columns = ["Dates"])
dates
```

```python
jan_to_mar = dates[(dates["Dates"] >= '2023-01-01') & (dates["Dates"] <= '2023-03-31')]
apr_to_jun = dates[(dates["Dates"] >= '2023-04-01') & (dates["Dates"] <= '2023-06-30')]
jul_to_sep = dates[(dates["Dates"] >= '2023-07-01') & (dates["Dates"] <= '2023-09-30')]
oct_to_dec = dates[(dates["Dates"] >= '2023-10-01') & (dates["Dates"] <= '2023-12-31')]
```
