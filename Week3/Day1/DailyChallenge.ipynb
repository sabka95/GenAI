# **1. Data Preparation:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!pip install faker
```

```python
import calendar
from faker import Faker
```

```python
fake = Faker()
```

```python
cities = [fake.city() for i in range(10)]
```

```python
temperature = np.random.uniform(-5, 35, 120)
temperature = temperature.reshape(10, 12)
df = pd.DataFrame(temperature, index = cities, columns = calendar.month_abbr[1:])
df
```

# **2. Data Analysis:**

```python
annual_average_temp = df.mean(axis = 1)
annual_average_temp
```

```python
lowest_average_temp = annual_average_temp.idxmin()
lowest_average_temp
```

```python
highest_average_temp = annual_average_temp.idxmax()
highest_average_temp
```

# **3. Data Visualization:**

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df, cmap="coolwarm", annot=True)
plt.title("Monthly Temperature Trends for Each City")
plt.xlabel("Months")
plt.ylabel("Cities")
plt.show()
```

# **4. Report:**

Millerport records the lowest average temperature, while Jonathanborough has the highest.
