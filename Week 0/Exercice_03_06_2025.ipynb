```python
human_years = int(input("Enter the number of human years: "))

if human_years == 1 :
    cat_years = 15
    dog_years = 15
    liste = [human_years, cat_years, dog_years]
    print(liste)

elif human_years == 2 : 
    cat_years = 24
    dog_years = 24
    liste = [human_years, cat_years, dog_years]
    print(liste)

else : 
    cat_years = 24 + ((human_years - 2) * 4)
    dog_years = 24 + ((human_years - 2) * 5)
    liste = [human_years, cat_years, dog_years]
    print(liste)
```
